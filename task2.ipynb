```python
from string import punctuation

text = open('text.txt')

def count(text):
    dic = dict()
    count = 0
    for line in text:
        word = line.split()
        for item in word:
            count += 1
            item = item.strip(punctuation)
            word_lowercase = item.lower()
            if word_lowercase in dic:
                dic[word_lowercase] += 1
            else:
                dic[word_lowercase] = 1
#     print('the total number of words in the book is ',count) 
    return sorted(dic.items(),key = lambda items:items[1],reverse = True)
```

```python
count(text)
```
